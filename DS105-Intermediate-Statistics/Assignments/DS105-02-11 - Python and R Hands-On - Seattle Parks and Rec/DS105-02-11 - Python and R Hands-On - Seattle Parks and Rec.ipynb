---
# VARIABLE: `# of trips Winter`
<mark style="background: #ABF7F7A6;">positively skewed</mark><br>

````python
sns.displot(data = parksrec, x = '# of trips Winter', kde = True)
````

### Transformation: Square Root

````python
parksrec['# of trips Winter SQRT'] = np.sqrt(parksrec['# of trips Winter'])
````

````python
sns.displot(data = parksrec, x = '# of trips Winter SQRT', kde = True)
````


<span style="color:red">
    <h3>Transformation: Logarithmic</h></span>
   
Not a good transformation; transforms the data too far to the right.

````python
parksrec['# of trips Winter LOG'] = np.log(parksrec['# of trips Winter'])
````

````python
sns.displot(data = parksrec, x = '# of trips Winter LOG', kde = True)
````

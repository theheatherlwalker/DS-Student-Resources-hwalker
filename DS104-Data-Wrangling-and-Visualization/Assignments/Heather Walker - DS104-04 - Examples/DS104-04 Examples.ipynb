## DS104-04-03 - Histograms in Python

```python
# import `pandas` package
import pandas as pd
```

```python
# import dataset
HR = pd.read_csv(r'data/HR_data.csv')
```

### Histograms with Pandas

```python
# create a histogram of the data with the `pandas` package
HR.satisfaction_level.hist()
```

### Histograms with Seaborn

```python
# import `seaborn` package for plotting
import seaborn as sns
```

```python
# plot with `displot` from seaborn package 
# creates a histogram with a best-fit curved line:
sns.distplot(HR['satisfaction_level'])
```

### Multiple histograms with Seaborn

```python
# plot multiple histograms with `pairplot`
sns.pairplot(HR)
```

### Histograms with Matplotlib

```python
# import `matplotlib`
import matplotlib.mlab as mlab
```

```python
# import related subpackage
import matplotlib.pyplot as plt
```

```python
# create a variable for number of bins:
num_bins = 5
```

```python
# specify details of the plot
n, bins, patches = plt.hist(HR['satisfaction_level'], num_bins, facecolor='blue', alpha=.05)
```

```python
# function to show plot
# NOTE: in Jupyter Notebooks, I did not need this to see the plot
plt.show()
```

```python
# change alpha argument
n, bins, patches = plt.hist(HR['satisfaction_level'], num_bins, facecolor='blue', alpha=1)
```

```python
# add labels
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
```

```python
# add title
plt.title('Histogram of Satisfaction Level')
```

```python
# combine for entire plot:
num_bins = 5
n, bins, patches = plt.hist(HR['satisfaction_level'], num_bins, facecolor='blue', alpha=1)
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.title('Histogram of Satisfaction Level')
plt.show()
```

## DS104-04-06 - Bar Charts in Python

```python
# plot with function `.value_counts()`
HR['salary'].value_counts().plot(kind='bar')
```

```python
# adding title and labels
SalaryFreq = HR['salary'].value_counts().plot(kind='bar')
SalaryFreq.set_title("Salary Frequency")
SalaryFreq.set_xlabel("Salary Categories")
SalaryFreq.set_ylabel("Frequency")
```

```python
## DS104-04-08 - Stacked Bar Charts in Python
crosstab_df = pd.crosstab(HR['sales'], HR['salary'])
crosstab_df.plot.bar(stacked=True)
```

## DS104-04-11 - Scatterplots in Python

```python
# import dataset
nutrition = pd.read_csv(r'data/USDA_nutrition.csv')
```

```python
# create scatterplot with `pandas`
Plot = nutrition.plot.scatter(x='fats', y='sat fats')
Plot.set_title("Fats by Saturated Fats")
Plot.set_xlabel("Fats")
Plot.set_ylabel("Saturated Fats")
```

```python
# plot with 3rd quantitative variable
Plot = nutrition.plot.scatter(x='fats', y='sat fats', c='protein')
Plot.set_title("Fats by Saturated Fats and Protein")
Plot.set_xlabel("Fats")
Plot.set_ylabel("Saturated Fats")
```

```python
# adding color
Plot = nutrition.plot.scatter(x='fats', y='sat fats', c='protein', cmap='coolwarm')
Plot.set_title("Fats by Saturated Fats and Protein")
Plot.set_xlabel("Fats")
Plot.set_ylabel("Saturated Fats")
```

```python
# display by size
Plot = nutrition.plot.scatter(x='fats', y='sat fats', s=nutrition['protein'])
Plot.set_title("Fats by Saturated Fats and Protein")
Plot.set_xlabel("Fats")
Plot.set_ylabel("Saturated Fats")
```

```python
# increase the size of the points
Plot = nutrition.plot.scatter(x='fats', y='sat fats', s=nutrition['protein']*2)
Plot.set_title("Fats by Saturated Fats and Protein")
Plot.set_xlabel("Fats")
Plot.set_ylabel("Saturated Fats")
```

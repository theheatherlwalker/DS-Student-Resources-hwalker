```python
# import needed packages
import pandas as pd
import seaborn as sns

# import `matplotlib`
import matplotlib.mlab as mlab

# import related subpackage
import matplotlib.pyplot as plt
```

## Part 2

```python
# import dataset
carmakes = pd.read_csv(r'data/L3P2.csv')
```

```python
# adding title and labels
carmakesPlot = carmakes.value_counts().plot(kind='bar')
carmakesPlot.set_title("Count of Car Makes")
carmakesPlot.set_xlabel("Car Make")
carmakesPlot.set_ylabel("Count")
```

## Part 5

```python
# import dataset
heartattacks = pd.read_csv(r'data/L5P5.csv')
```

```python
# combine columns to a date
heartattacks['Date'] = heartattacks["Month"] + " " + heartattacks["Day"] + " " + heartattacks["Year"].map(str)
```

```python
# view the dataframe
heartattacks.head()
```

```python
# create line plot with title and labels
plt.plot(heartattacks['Date'], heartattacks['Heart Attacks'])
plt.xlabel('Date')
plt.ylabel('# of Heart Attacks')
plt.title("Heart Attacks, May through July 2003")
```

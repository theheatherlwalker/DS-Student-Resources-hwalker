```python
# import needed packages
import pandas as pd
```

```python
# import dataset
FakeNews = pd.read_excel(r'data/FakeNews.xlsx')
```

```python
# view first rows of dataframe
FakeNews.head()
```

---

```python
# 1. Add a column labeled `StoryType` and fill it with `Fake`
FakeNews['StoryType'] = 'Fake'
```

```python
# view changes to dataframe
FakeNews.head()
```

---

```python
# 2. Remove the `when` column
FakeNews2 = FakeNews.drop(['when'], axis=1)
```

```python
# view result
FakeNews2.head()
```

---

```python
# 3. Separate the `url` column to a website column and domain column
FakeNewsUrl = FakeNews2['url'].str.split('_/', expand=True).rename(columns = lambda x: "URL"+str(x+1))
```

```python
# view result
FakeNewsUrl.head()
```

```python
# add the split url dataframe back to main dataframe
FakeNews3 = pd.concat([FakeNews2, FakeNewsUrl], axis=1)
```

```python
# view result
FakeNews3.head()
```

```python
# remove original `url` column
FakeNews3 = FakeNews3.drop(['url'], axis=1)
```

```python
# view result
FakeNews3.head()
```

---

```python
# 4. Put back together the domain column.
FakeNews3['FullUrl'] = FakeNews3["URL1"] + "_/ " + FakeNews3["URL2"].map(str) 
```

```python
# view result
FakeNews3.head()
```

```python
# drop `URL1` and `URL2` columns
FakeNews3.drop(['URL1', 'URL2'], axis=1, inplace=True)
```

```python
# view result
FakeNews3.head()
```

---

```python
# 5. Keep only the first 10 rows of data
FakeNews4 = FakeNews3[:10]
```

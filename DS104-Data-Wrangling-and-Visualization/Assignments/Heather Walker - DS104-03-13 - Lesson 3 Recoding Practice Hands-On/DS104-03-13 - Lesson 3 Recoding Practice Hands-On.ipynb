## Part 1: Recoding in Python

```python
# import needed packages
import pandas as pd
```

```python
# import dataset
eatingHabits = pd.read_csv(r'data/Eating_Habits.csv')
```

```python
# view dataset
eatingHabits.head()
```

### 1.1  Recode `Activity` into a new variable.
`0` should be `not eating chocolate`, and `1` should be `eating chocolate`.

```python
# find unique values to confirm syntax of entry
eatingHabits.Activity.unique()
```

```python
# create copy of dataset
eatingHabits1 = eatingHabits
```

```python
# create a function to recode `Activity`
def activity (series): 
    if series == 'Eating candy, chocolate bars':
        return 1
    else: 
        return 0
```

```python
# recode the "Activity" column with function `activity`
eatingHabits1['ActivityR'] = eatingHabits['Activity'].apply(activity)
```

```python
# view result
eatingHabits1.head()
```

---
### 1.2  Recode `Frequency` from text to numbers in the same variable. 
`0` should be the lowest frequency.

```python
# find unique values to confirm syntax of entry
eatingHabits1.Frequency.unique()
```

```python
# definition for changing text to numbers
frequencyNumbers = {"Frequency": {'Never': 0, 
                                  'Seldom': 1, 
                                  'At least once a week': 2, 
                                  'Once a day': 3, 
                                  'More than once a day': 4}}
```

```python
# replace text with numbers in same variable
eatingHabits1.replace(frequencyNumbers, inplace=True)
```

```python
eatingHabits1.head()
```

### 1.3  Recode `Sex` from text to numbers in the same variable. 

```python
# find unique values to confirm syntax of entry
eatingHabits1.Sex.unique()
```

### 1.4  Dummy code the `Age` group variable.

```python
# find unique values
eatingHabits1['Age group'].unique()
```

```python
# generate data frame with dummies
ageGroupDummy = pd.get_dummies(eatingHabits1['Age group'])
```

```python
# concatenate the dummies data frame to original in new variable
eatingHabits2 = pd.concat([eatingHabits1, ageGroupDummy],axis=1)
```

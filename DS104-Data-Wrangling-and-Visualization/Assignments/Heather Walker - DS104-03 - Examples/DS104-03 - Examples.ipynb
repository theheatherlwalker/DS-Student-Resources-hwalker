## DS104-03-04 - Recoding into a new variable in Python

```python
# import needed packages
import pandas as pd
```

```python
# import dataset
superheroes = pd.read_csv(r'data/superheroes.csv')
```

```python
# view dataset
superheroes.head()
```

```python
# create a function named `gender`
def gender (series): 
    if series == "Male":
        return 0
    if series == "Female": 
        return 1
```

```python
# create copy of dataset
superheroes1 = superheroes
```

```python
# view the copy
superheroes1.head()
```

```python
# recode the "Gender" column with function `gender`
superheroes1['GenderR'] = superheroes1['Gender'].apply(gender)
```

```python
superheroes1.head()
```

## DS104-03-05 - Grouping with a Recode in Python

```python
# function to recode into 3 groups: Comic, Screen, Book
def media (series): 
    if series == "Marvel Comics" : 
        return "Comic"
    if series == "DC Comics": 
        return "Comic"
    if series == "NBC - Heroes" : 
        return "Screen"
    if series == "Dark Horse Comics" : 
        return "Comic"
    if series == "George Lucas" : 
        return "Screen"
    if series == "Image Comics" : 
        return "Comic"
    if series == "HarperCollins" : 
        return "Book"
    if series == "Star Trek" : 
        return "Screen"
    if series == "SyFy" : 
        return "Screen"
    if series == "Team Epic TV" : 
        return "Screen"
    if series == "IDW Publishing" : 
        return "Book"
    if series == "ABC Studios" : 
        return "Screen"
    if series == "Shueisha" : 
        return "Unknown"
    if series == "Icon Comics": 
        return "Comic"
    if series == "Wildstorm" : 
        return "Unknown"
    if series == "Sony Pictures" : 
        return "Screen"
    if series == "South Park" : 
        return "Screen"
    if series == "J.R.R. Tolkein" : 
        return "Book"
    if series == "Universal Studios" : 
        return "Screen"
    if series == "Rebellion" : 
        return "Unknown"
    if series == "Titan Books" : 
        return "Book"
    if series == "Hanna-Barbera" : 
        return "Screen"
    if series == "Microsoft" : 
        return "Unknown"
    if series == "J.K. Rowling" : 
        return "Book"
```

```python
# add a column `Media` and us `media` function to recode
superheroes1['Media'] = superheroes1['Publisher'].apply(media)
```

```python
superheroes1.head()
```

## DS104-03-06 - Recoding from Continuous to Categorical in Python

```python
# function to recode continuous variable `Height` into categories
# NOTE: `Height` is in cm
def height_recode (series): 
    if series < 152.4:
        return 0
    if series >= 152.4: 
        return 1
```

```python
# add column with recoded categorical height
superheroes1['Height_5ftYN'] = superheroes1['Height'].apply(height_recode)
```

```python
# view result
superheroes1.head()
```
